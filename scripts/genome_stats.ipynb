```python
%matplotlib inline
from Bio import SeqIO
import numpy as np
import matplotlib.pyplot as plt

len_total = 0
chr_name = []
chr_len = []
nuc_cnt = {'A':0, 'C':0, 'G':0, 'T':0}

recs = SeqIO.parse('../build/ref_genome.fa', 'fasta')

for r in recs:
    chr_name.append(r.id)
    l = len(r)
    chr_len.append(l)
    len_total += l
    for nuc in "ACGT":
        nuc_cnt[nuc] += r.seq.count(nuc)

```

```python
# plot chromosome lengths
y_pos = np.arange(len(chr_name))
plt.figure(num=None, figsize=(8,2), dpi=80, facecolor='w', edgecolor='k')
plt.barh(y_pos, chr_len, align='center', height=0.5, alpha=0.5)
plt.yticks(y_pos, chr_name)
plt.xlabel('length')
plt.title('Chromosome lengths')
plt.show()
```

```python
# plot base frequencies
lbl_nuc = ['A', 'C', 'G', 'T']
y_pos = np.arange(len(lbl_nuc))
nuc_freq = [ nuc_cnt[n]/len_total for n in lbl_nuc ]

plt.figure(num=None, figsize=(2,8), dpi=80, facecolor='w', edgecolor='k')
plt.bar(y_pos, nuc_freq, align='center', color='g', alpha=0.5)
plt.xticks(y_pos, lbl_nuc)
plt.ylabel('freq')
plt.title('Nucleotide frequencies')
plt.show()
```
